```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([23,34,42,49,55,61,68,73,78,82])        # create a numpy array
y = np.array([25,29,33,37,44,47,57,62,67,71])
z = np.array([87,82,78,73,67,63,55,51,47,39])      
```

### Line Plot

```python
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,y,marker = 'o', mfc = 'red', color = 'green' ,ls ='--', ms = 8 ,label = '2023');
plt.plot(x,z,marker = 'o', mfc = 'blue', color = 'green' ,ls ='--', ms = 8,label = '2024');
plt.grid()                  # Marker Style=> >,<,^,H,o,*
plt.legend();
```

### Box Plot

```python
marks = [67,65,55,59,69,72,51,70,53,98,23]
```

```python
plt.boxplot(marks);     # use to present outliers
```

```python
# calculate 75% and 25%

q3 = np.percentile(marks, 75)
q1 = np.percentile(marks, 25)
q3 , q1
```

```python
# caluclate IQR(inter portal range)
iqr = q3 - q1
```

```python
iqr
```

```python
# calculate upper bound and lower bound
# upper bound

ub = (q3 + (1.5 * iqr))

#lower bound
lb = (q1 - (1.5 * iqr))
```

```python
ub
```
