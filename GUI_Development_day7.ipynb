```python
import tkinter as tk
from tkinter import BooleanVar
from tkinter.ttk import *
from tkinter import  scrolledtext
from tkinter  import messagebox
```

```python
top = tk.Tk()  # create a window

### WIDGETS

top.title('My Apllication') # give the title to window

top.geometry('500x400')  # give the window size

# button
b1 = tk.Button(top,text='click here!!!',bg = 'orange').grid(column = 0 , row = 1) # use to create the button

# entry button
e1 = tk.Entry(top).grid(column = 1 , row = 1) # use to create entry button

# text button
t1 = tk.Text(top).grid(column = 2 , row = 2)

# label
l1 = tk.Label(top, text = 'welcome', font = ('Ariel Bold',50)).grid(column = 2 , row = 3)

# combobox
combo = Combobox(top)
combo['values'] = [ 1, 2, 3, 'next' ]
combo.current(2)
combo.grid(column = 0 , row = 0)

# check button
ch = BooleanVar()
ch.set(True)
cb = Checkbutton(top, text = 'Done' , var = ch, width = 30).grid(column = 0 , row = 0)

# scrolled text
st = scrolledtext.ScrolledText(top, width = 40, height = 10).grid(column = 3, row = 3)            # use to scroll the widget.

top.mainloop()    # hold that window
```

```python
# square of a number

top = tk.Tk()

top.title('Square') 
top.geometry('500x400')

a = tk.IntVar()
def square():
    num = a.get()
    sq = num * num
    result = tk.Label(top, text = 'square is %i:' % sq).place(x = 120 , y = 160)
    messagebox.showinfo("result","square is %i" %sq)

l = tk.Label(top , text = 'input').place(x = 20 , y = 100)  # place is use to give the place for the different widgets. 
e = tk.Entry(top, textvariable = a).place(x = 100 , y = 100)
b = tk.Button(top, text = 'submit' , bg = 'green', command = square).place(x = 130 , y = 150)


top.mainloop() 
```

```python

# Addition

top = tk.Tk()

top.title('operations') 
top.geometry('500x400')

# square
a1 = tk.IntVar()
a2 = tk.IntVar()

def addition():
    add = a1.get() + a2.get()
    messagebox.showinfo("result","addition is %i" %add) 

l = tk.Label(top , text = 'input').place(x = 20 , y = 100)# place is use to give the place for the different widgets. 
l1 = tk.Label(top , text = 'input').place(x = 20 , y = 130)
e = tk.Entry(top, textvariable = a1).place(x = 100 , y = 100)
e1 = tk.Entry(top, textvariable = a2).place(x = 100 , y = 130)

b = tk.Button(top, text = 'submit' , bg = 'green', command = addition, cursor='hand2').place(x = 130 , y = 150)

top.mainloop() 
```
