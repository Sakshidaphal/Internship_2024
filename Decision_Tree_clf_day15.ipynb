```python
#seperate the couumns
x = df[['Variance',	'Skewness',	'Curtosis',	'Entropy']]  # OR (   x = df.drop('Class',axis=1)   )if we have more than 100 columns
y = df['Class']
```

# Cross Validation

```python
# used to estimate the performance or accuracy of predictive data
# holdout method = use when you have a very large dataset
```

```python
from sklearn.model_selection import train_test_split
```

```python
# 4 variables will create
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , random_state = 0 , test_size = 0.25)
```

```python
x_train.shape         # training data
```

```python
y_train.shape          # training data  (it only shows one column because it is use for output)
```

### Build the Model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(random_state = 0)
```

```python
dt.fit(x_train , y_train)
```

```python
# predict on test data
```

```python
y_pred = dt.predict(x_test)    # x_test is input testing data
```

```python
y_pred
```

```python
y_test
```

```python
# create a dataframe for comparision
result = pd.DataFrame({
    'actual':y_test,
    'predict':y_pred
})
```

```python
result
```

### Model Evalution

```python
# confusion matrix

from sklearn.metrics import ConfusionMatrixDisplay
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
 # classification report
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
# overall accuracy
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
